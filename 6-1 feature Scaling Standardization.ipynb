# Standardization

## manual approach

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
total_bill = df['total_bill'].tolist()
```

```python
mean = np.mean(total_bill)
std = np.std(total_bill)
```

```python
mean, std
```

```python
normalized_data =[]

for i in total_bill:
    z_score = (i-mean)/std
    normalized_data.append(z_score)
    
    print(f'{len(normalized_data)} : The element is : {normalized_data} \n')
    
    
print( normalized_data)
```

```python
len(normalized_data)
```

```python
sns.histplot(total_bill)

plt.title('Before Standardization')
```

```python
plt.figure(figsize=(10,4))
sns.histplot(normalized_data)

plt.title('After Standardization')
```

```python
# Plot original data
sns.histplot(total_bill, color='blue', label='Original Data', kde=True)

# Plot normalized data
sns.histplot(normalized_data, color='red', label='Normalized Data', kde=True)

# Add labels and legend
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.title("Comparison of Original and Normalized Data")

# Show the plot
plt.show()
```
