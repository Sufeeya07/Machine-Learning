```python
from sklearn.datasets import load_iris
from sklearn.neighbors import LocalOutlierFactor
import matplotlib.pyplot as plt

# Load Iris dataset
iris = load_iris()
X = iris.data[:, :2]  # Take only 2 features for easy plotting
```

```python
n_neighbors=20: How many neighbors to use to estimate local density.
```

```python
lof= LocalOutlierFactor(n_neighbors=20) # k neighbour
lof
```

```python
y_pred=lof.fit_predict(X)
y_pred
```

```python
outliers=X[y_pred==-1]
inliers=X[y_pred==1]
```

```python
# Plot results
plt.scatter(inliers[:, 0], inliers[:, 1], label='Inliers', color='blue')
plt.scatter(outliers[:, 0], outliers[:, 1], label='Outliers', color='red')
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("LOF on Iris Dataset")
plt.legend()
plt.grid(True)
plt.show()
```
