# Decision Tree Regressor 

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import  fetch_california_housing
cal_df = fetch_california_housing()
```

```python
cal_df
```

```python
# indepndent and dependent fetaures

X=pd.DataFrame(cal_df.data, columns=cal_df.feature_names)
y=cal_df.target
```

```python
X, y
```

```python
# train test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
reg= DecisionTreeRegressor()
```

```python
reg.fit(X_train, y_train)
```

```python
# prediction

y_pred= reg.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
score= r2_score(y_test, y_pred)
print(score)
```

# Hyperparameter Tuning

```python
parameter={ 'criterion': ['squared_error', 'absolute_error', 'friedman_mse', 'poisson'],
           'splitter': ['best', 'random'],
           'max_depth': [1,2,3,4,5],
           'max_features':['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regcv= GridSearchCV(reg, param_grid=parameter, cv=5, scoring='neg_mean_squared_error')
```

```python
# train data
regcv.fit(X_train , y_train)
```

```python
regcv.best_params_
```

```python
y_pred=regcv.predict(X_test)
```

```python
r2_score(y_pred, y_test)
```
