# Data Interpolation

# 1-Linear Interpolation (staright line)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.array([1,2,3,4,5])
y= np.array([2,4,6,8,10])
```

```python
plt.scatter(x,y)
```

```python
x_new= np.linspace(1,5,10) # create new x values
y_inter= np.interp(x_new,x , y) # interpolation y value
print(y_inter)
```

```python
plt.scatter(x_new, y_inter)
```

```python
# cubic interpolation (scipy)
```

```python
x= np.array([1,2,3,4,5])
y= np.array([1,8,27,64,125]) # cube of x
```

```python
from scipy.interpolate import interp1d
```

```python
# create cubic interpolation

f= interp1d(x,y,kind='cubic')
```

```python
x_new = np.linspace(1,5,10)
y_interp= f(x_new)
print(y_interp)
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x_new, y_interp)
```

```python
# polynomial interpolation
```

```python
x= np.array([1,2,3,4,5])
y= np.array([1,4,9,16,25]) # sq of x 
```

```python
p = np.polyfit(x, y , 8) # fit a 2nd degree , how much do want to bend
```

```python
x_new = np.linspace(1,5,10)
y_interpol= np.polyval(p,x_new)
print(y_interp)
```

```python
plt.scatter(x_new, y_interpol)
```

```python
import pandas as pd
import numpy as np
```

```python
data ={'day':[1,2,3,4,5,6,7],
      'temp': [30, np.nan, 32, np.nan, np.nan, 35, 36]}

df= pd.DataFrame(data)
print('Before interpolation :\n ' , df)
```

```python
plt.scatter(df['day'], df['temp'])
```

```python
df['temp_interpol'] = df['temp'].interpolate(method='linear')

print('\nAfter interpolation: \n' , df)
```

```python
plt.plot(df['day'], df['temp_interpol'], linestyle='--', marker='*', color='blue', label="Interpolated Data")
plt.show()
```

```python
data = {'Day': np.arange(1, 11),
        'Temperature': [30, np.nan, 28, np.nan, 25, np.nan, np.nan, 29, np.nan, 34]}

df = pd.DataFrame(data)
print('Before interpolation :\n ' , df)
```

```python
df['temp_cubic'] = df['Temperature'].interpolate(method='cubic')

print('\nAfter interpolation: \n' , df)
```

```python
# Original points
plt.scatter(df['Day'], df['Temperature'], color='red', label="Original Data (With NaN)", zorder=2)
```

```python
plt.plot(df['Day'], df['temp_cubic'], linestyle='--', marker='o', color='green', label="Cubic Interpolated ", zorder=1)

plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.title("Cubic Interpolation - Smooth Curve")
plt.legend()
plt.grid(True)

plt.show()
```
